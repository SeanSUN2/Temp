## Part 2: Matplotlib 绘图
绘制函数 $f(x) = x^2$ 及其在 $x=0$ 和 $x=8$ 处的切线。

```python
def f(x):
    return x**2

def f_deriv(x):
    return 2 * x

x = np.linspace(-15, 15, 400)
plt.figure(figsize=(6, 6))
plt.plot(x, f(x), label='f(x) = x²')
plt.plot(x, f_deriv(0) * x, label='Tangent at x=0')
plt.plot(x, f_deriv(8) * (x - 8) + f(8), label='Tangent at x=8')
plt.ylim(-100, 300)
plt.legend()
plt.title("Function f(x) and its tangents")
plt.show()
```
